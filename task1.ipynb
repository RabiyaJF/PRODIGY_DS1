```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

DATASET : https://www.kaggle.com/datasets/wenruliu/adult-income-dataset

```python
df=pd.read_csv('adult.csv')
df.head()
```

```python
df.info()
```

```python
df.isin(['?']).sum()
```

```python
df.replace(to_replace='?',value=np.nan,inplace=True)
```

```python
df.isin(['?']).sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['age'],bins=10,edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
object_columns = df.select_dtypes(include=['object']).columns
print("Object columns:", object_columns)
```

```python
vals = df['workclass'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(vals.index,vals,color='green',edgecolor='black')
plt.title('Distribution of workclass')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.show()
```

```python
vals = df['education'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(vals.index,vals,edgecolor='black',color='#0D0AB0')
plt.title('Distribution of education')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['hours-per-week'],bins=20,color='#6A0DAD')
plt.title('Distribution of Hours per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Frequency')
plt.show()
```

```python
vals = df['gender'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(vals.index,vals,edgecolor='black',color='#D00ACE')
plt.title('Distribution of gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
vals = df['native-country'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(vals.index,vals,edgecolor='black',color='#0FADA1')
plt.title('Distribution of native-country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
x=df['native-country']
y=df['age']
c = np.arange(len(x)) 
plt.figure(figsize=(10,6))
plt.scatter(x,y,c=c,cmap='coolwarm')
plt.xlabel('Country')
plt.ylabel('Age')
plt.xticks(rotation=90)
plt.show()
```
